```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from skimage.exposure import equalize_adapthist

# --- 1. Custom CLAHE Preprocessing Function ---
def clahe_enhancement(img_array):
    """
    Applies CLAHE (Contrast Limited Adaptive Histogram Equalization)
    to a single-channel (grayscale) image array.
    
    The Keras generator passes the image as a (H, W, C) array.
    Since we are using color_mode='grayscale', C=1.
    """
    # Remove the channel dimension for skimage processing: (H, W, 1) -> (H, W)
    if img_array.ndim == 3 and img_array.shape[-1] == 1:
        img_2d = img_array[:, :, 0]
    else:
        # Should not happen if color_mode='grayscale' is set
        img_2d = img_array

    # The input image is already normalized (0 to 1) due to rescale=1./255.
    # CLAHE function in skimage works best on floating point data in [0, 1] range.
    
    # Apply CLAHE
    img_enhanced = equalize_adapthist(img_2d, clip_limit=0.03)
    
    # Add the channel dimension back: (H, W) -> (H, W, 1)
    # Note: We return the image in the [0, 1] float range.
    return img_enhanced[..., np.newaxis]

# The rescale=1./255 step is automatically handled by the generator
# and happens *before* this preprocessing function.
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory(r"dataset/train_set",
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```
